```python
from collections import deque
```

```python
#graph representation
graph= {
    'A': ['B','C'],
    'B':['D','E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}
```

```python
print(graph)
```

```python
start='A'
goal='F'
```

```python
queue=deque([start])
visited=set([start])
parent=()
step=0
```

```python
print(queue)
```

```python
parent = {start : None}
print (parent)
```

```python
while queue :
  print("current queue: ", list(queue))
  # Removed input("press enter for next step..") for continuous execution
  step += 1
  node = queue.popleft()

  print("\n====")
  print("STEP: ",step)
  print("Popped node: ",node)
  print("Queue after pop: ", list(queue))

  if node == goal:
    print("goal found! stop bfs ")
    break
```

```python
for neighbour in graph[node]:
  print(f" checking neighbor {neighbor}of {node}...")
  print("now processing neighbor : ",neighbor)
  if neighbor not in visited:

    visited.add(neighbor)
    parent[neighbor]=node
    queue.append(neighbor)

    print(" parent[( )]={}",format(neighbor,node))
    print("added to queue: ",neighbor)

  else:
    print("already")
    #print("current queue: ",list[queue])

```
