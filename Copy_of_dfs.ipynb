```python
import matplotlib.pyplot as plt

# Your tree/graph (adjacency list)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

# 1) Give each node a position (x, y)
pos = {
    'A': (0, 2),
    'B': (-1, 1),
    'C': ( 1, 1),
    'D': (-1.5, 0),
    'E': (-0.5, 0),
    'F': ( 1.0, 0),
}

plt.figure(figsize=(7,4))

# 2) Draw edges (lines)
for u in graph:
    x1, y1 = pos[u]
    for v in graph[u]:
        x2, y2 = pos[v]
        plt.plot([x1, x2], [y1, y2])  # simple line

# 3) Draw nodes (points) + labels
for n, (x, y) in pos.items():
    plt.scatter(x, y, s=600)
    plt.text(x, y, n, ha='center', va='center', fontsize=14)

plt.title("Tree Visualization using Matplotlib")
plt.axis('off')
plt.show()
```

```python
stack=[start]
visited= set ([start])
parent ={start:None}
step =0

print("Initial stack:", list(stack))
print("Initial visited:", visited)
print("Initial parent :", parent)
while stack:
  input("Press enter for next step...")
  step += 1
  node = stack.pop()
  print("\nSTEP", step)
  print("Popped node:", node)
  print("stack after pop:", list(stack))
  if node == goal:
    print("Goal found! Stop BFS.")
    break
  for neighbor in(graph[node]):
    print(neighbor)
    if neighbor not in visited:
      visited.add(neighbor)
      parent[neighbor]=node
      stack.append(neighbor)
      print("  Added to visited:", neighbor)
      print(f"  parent[{neighbor}] = {node}")
      print("  Added to stack:", neighbor)
    print("stack now:", list(stack))
    print("Visited now:", visited)

print("\nBFS ended.")
print("Final visited:", visited)
print("Final parent :", parent)
```

```python

def reconstruct_path(parent, start, goal):

    if goal not in parent:
        return None

    path = []

    current = goal


    while current is not None:
        path.append(current)
        current = parent[current]

    path.reverse()
    return path
```

```python
path = reconstruct_path(parent, start, goal)

print("\nShortest path from", start, "to", goal, ":")
print(path)
```
