```python
#graph representation
graph= {
    'A': ['B','C'],
    'B':['D','E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}
```

```python
print(graph)
```

```python
start='A'
goal='F'
```

```python
stack=([start])
type(stack)
visited=set([start])
parent=()
step=0
```

```python
print("initial stack:",stack)
print("initial visited:",visited)
print("initial parent:",parent)
```

```python
parent = {start : None}
print (parent)
```

```python
while stack :
  # input("press enter for next BFS step...") #input for continuous execution
  print("current stack: ", list(stack))

  step += 1
  node = stack.pop()

  print("\n====")
  print("STEP: ",step)
  print("Popped node: ",node)
  print("stack after pop: ", list(stack))

  if node == goal:
    print("goal found! stop DFS")
    break
```

```python
for neighbor in graph[node]:
    print(f"checking neighbor {neighbor} of {node}...")
    if neighbor not in visited:
        visited.add(neighbor)
        # Ensure parent is a dictionary to allow assignment
        if not isinstance(parent, dict):
            parent = {}
        parent[neighbor] = node
        stack.append(neighbor)
        print("Added to visited:", neighbor)
        print(f"parent[{neighbor}] = {node}")
        print("Added to stack:", neighbor)
        print("stack now:", list(stack))
        print("visited now:", visited)

print("\n DFS Iteration finished.")
print("Final visited for this step:", visited)
print("Final parent mapping:", parent)
```
